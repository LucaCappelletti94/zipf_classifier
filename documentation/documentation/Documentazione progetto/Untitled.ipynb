```python
import os
from tqdm import tqdm_notebook as tqdm
import re
```

```python
def clear_filename(filename:str):
    target = " "
    root, stem = "/".join(filename.split("/")[:-1]), filename.split("/")[-1]
    new_name = root +"/"+ re.sub(" ", "_", stem)
    os.rename(filename, new_name)
```

```python
def clear_filename(filename:str):
    target = "_-_"
    new_name = re.sub(target, "_", filename)
    os.rename(filename, new_name)
```

```python
[clear_filename("{path}/{file}".format(path=path, file=file)) for path, dirs, files in tqdm(os.walk("./")) for file in files]
```
